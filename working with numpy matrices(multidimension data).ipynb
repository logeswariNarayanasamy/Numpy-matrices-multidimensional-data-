```python
import numpy as np
from scipy.stats import skew,kurtosis
```

```python
def describe_stats(weights):
    mean = np.mean(weights)
    median = np.median(weights)
    std_dev = np.std(weights)
    skewness = skew(weights)
    kurt = kurtosis(weights)
    return mean, median, std_dev, skewness, kurt

```

```python
male_weights = [60,70,80,90]
female_weights = [50,55,65,75]
```

```python
female_stats = describe_stats(female_weights)
male_stats = describe_stats(male_weights)

```

```python
print("Female Weight Stats (Mean, Median, Std Dev, Skewness, Kurtosis):", female_stats)
print("Male Weight Stats (Mean, Median, Std Dev, Skewness, Kurtosis):",male_stats)
```
